```python
import pandas as pd
from collections import defaultdict
import os
```

```python
df1 = pd.read_csv("G:/companyListShortenned.csv")
```

```python
df1
```

```python
sect = defaultdict(str)
with open("G:/files.txt") as file:
    for line in file:
        data = line.strip()
        row = df1.loc[df1['Symbol'] == data]
        category = str(row['Sector'].values.tolist()).replace("['",'').replace("']",'')
        #print(category)
        if category not in sect:
            sect[category]=1
        else:
            sect[category] +=1
#print (str(sect))

print('{:20}{:10}{:10}'.format("Category", "Number", "Percent"))
for key in sorted(sect.keys()):
    print('{:20} {:10} {:2.2f}'.format(key, sect[key], (sect[key]/2941)*100))
print('{:20} {:10} {:10}'.format("Total", "2941", ""))

#for category, count in sect.items():
#    print('{} {} {}'.format(category, count,count/2941))
    
```
